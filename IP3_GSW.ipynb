### Question 4

```python
car_sales = pd.read_csv('car-sales.csv')


car_sales.info()
car_sales.head()
```

### Question 5

```python
car_sales['Price'] = car_sales['Price'].replace({'\$': '', ',': ''}, regex=True).astype('int64')
car_sales.head()
```

### Question 6

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

### Question 7

```python
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))
car_sales['Sale Date']=car_sales['Sale Date'].dt.date
car_sales.head()
```

### Question 8

```python
# Line Plot: Relationship between 'Price' and 'Total Sales'
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title("Price vs Total Sales Line Plot")
plt.xlabel("Price")
plt.ylabel("Total Sales")
```

### Question 9

```python
plt.scatter(x=car_sales["Odometer (KM)"], y=car_sales["Price"])
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.title("Odometer vs Price Scatter Plot")
```

### Question 10

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.title('Make vs Odometer (KM) Bar Plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

### Question 12

```python
#Question 12
fig,axes = plt.subplots (2,2, figsize=(10,10))

axes[0,0].plot(car_sales['Price'], car_sales['Total Sales'])
axes[0,0].set_title('Price vs Total Sales Line Plot')

sns.scatterplot(x=car_sales['Odometer (KM)'], y=car_sales['Price'],ax=axes[0,1])
axes[0,1].set_title('Odometer vs Price Scatterplot')

sns.barplot(x=car_sales['Make'], y=car_sales['Odometer (KM)'], errorbar=None, ax=axes[1,0])
axes[1,0].set_title('Make vs Odometer Barplot')

sns.countplot(x=car_sales['Make'], ax=axes[1,1])
axes[1,1].set_title('Histogram of Car Make')
```

### Question 16

```python
corr_matrix=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
#corr_matrix
sns.heatmap(corr_matrix, annot=True)
```

Price and Odometer (KM) have a negative correlation. So, as the car's odometer rises the price of the car will fall.
Total Sales and Odometer (KM) also have a negative correlation. This means that as cars odometers rise the total sales of those cars will begin to fall.
